```python
from bs4 import BeautifulSoup
import requests
import re
import numpy as np
import pandas as pd
```

```python
box_office_url = "https://www.boxofficemojo.com/chart/top_lifetime_gross/?area=XWW"
top_charts_page = requests.get(box_office_url)
soup_top_charts_page = BeautifulSoup(top_charts_page.content, 'html.parser')

all_movies = soup_top_charts_page.find_all("a", {"class": "a-link-normal"})
all_movie_titles = []

for movie in all_movies:
    movie_string = str(movie)
    if "title" in movie_string:
        all_movie_titles.append(movie.text)
        
all_movie_titles
```
